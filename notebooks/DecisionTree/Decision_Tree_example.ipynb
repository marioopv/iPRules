# Decision Tree

## Load Dataset

```python

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

#Load Dataset
dataset = load_iris()

print(dataset.feature_names)
#Define dataset
X_train, X_test, y_train, y_test = train_test_split(dataset.data, dataset.target, test_size=0.1, random_state=1)
```

## Define Random Forest

```python
from sklearn.tree import DecisionTreeClassifier

# Define scorer
tree = DecisionTreeClassifier(random_state=0)
```

## Fit

```python
# Train the random forest classifier on the Iris dataset
tree.fit(X_train, y_train)

# Make predictions for the test set
y_pred_test = tree.predict(X_test)
```

## Print Model

```python
from sklearn.tree import export_text
from sklearn.tree import export_text, plot_tree, export_graphviz
import matplotlib.pyplot as plt


r = export_text(tree, feature_names=list(dataset.feature_names))
#print(f'Tree :\n{r}\n')
plot_tree(tree, filled=True)
plt.show()
```
